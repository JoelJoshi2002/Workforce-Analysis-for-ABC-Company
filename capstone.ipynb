```python
# How Many Are There In Each Team and Precentage splitting with respect to the total employees.
```

```python
team_distribution = df['Team'].value_counts();
```

```python
team_distribution
```

```python
team_percentage = (team_distribution / len(df)) * 100
```

```python
team_percentage
```

```python
team_percentage.sum()
```

```python
plt.figure(figsize=(10, 10))
team_percentage.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='tab20', pctdistance=0.85)
plt.title('Percentage Distribution of Employees Across Teams')
plt.ylabel('') 
plt.show()
```

```python
# Visualization: Team Distribution
```

```python
team_distribution.plot(kind='bar', color='skyblue', title='Employee Distribution by Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

```python
bins = [20, 25, 30, 35]
age_groups = pd.cut(df['Age'], bins=bins, labels=["20-25", "26-30", "31-35"], right=False)
age_group_distribution = age_groups.value_counts()
```

```python
age_group_distribution
```

```python
age_group_distribution.plot(kind='bar', color='coral', title='Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# 4
```

```python
salary_by_team = df.groupby('Team')['Salary'].sum()
salary_by_position = df.groupby('Position')['Salary'].sum()
```

```python
salary_by_team
```

```python
salary_by_position
```

```python
highest_salary_team = salary_by_team.idxmax()
highest_salary_position = salary_by_position.idxmax()
```

```python
highest_salary_team
```

```python
highest_salary_position
```

```python
print(f"\nTeam with Highest Salary Expenditure: {highest_salary_team} with {salary_by_team.max()} salary")
print(f"Position with Highest Salary Expenditure: {highest_salary_position} with {salary_by_position.max()} salary")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
salary_by_team.plot(kind='bar', color='teal', ax=ax[0], title='Salary by Team')
ax[0].set_ylabel('Total Salary')
ax[0].set_xlabel('Team')

salary_by_position.plot(kind='bar', color='purple', ax=ax[1], title='Salary by Position')
ax[1].set_ylabel('Total Salary')
ax[1].set_xlabel('Position')

plt.tight_layout()
plt.show()
```

```python
Note: Team with Highest Salary Expenditure: Cleveland Cavaliers 
Position with Highest Salary Expenditure: C 
```

```python
correlation = df['Age'].corr(df['Salary'])
print(f"\nCorrelation between Age and Salary: {correlation:.2f}")
```

```python
sns.regplot(x='Age', y='Salary', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title(f'Correlation between Age and Salary: {correlation:.2f}')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Observations: The correlation coefficient 0.21 indicates a weak positive correlation between age and salary. 
As age increases, there is a slight upward trend in salary, but it is not very strong or significant.
```
